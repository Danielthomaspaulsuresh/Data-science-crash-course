# NUMBER GUESSING GAME LEVEL 2



## Guess the correct number with the maximum of seven tries using for loop


```python
secret_number = 56

for i in range(0,7):
    user_id = int(input("enter your guess = "))
    if secret_number == user_id:
        print("you guess is correct")
        break
    else:
        print("your guess is wrong")
```

# Guess the correct number with the maximum of seven tries using while loop

```python
secret_number = 56

while True:
    user_id = int(input("enter your guess = "))
    if secret_number == user_id:
        print("you guess is correct")
        break
    else:
        print("your guess is wrong")
```

## giving multiple condition

1. user should enter the number between 1 to 100
2. if the user number greater than the secrete than the program should say your number is greater than the secret number
3. if the user number lesser than the secrete than the program should say your number is lesser than the secret number

```python
print(" welcome to number guessing game")
print("Guess the number your Lucky!! hint: The number falls between 1 to 100")
secret_number = 65
for i in range(0,7):
    user_input = int(input("Enter your guess = "))
                     
    if user_input <1 or user_input >100:
        print("invalid number please enter the number between 1 to 100")

    elif user_input > secret_number:
        print("your number is higher")
    elif user_input < secret_number:
        print("your number is loweer")

    else:
        print("your guessed number is correct ")
        print(f"congratulation, You have guessed the number", {secret_number}, "in", {i}, "attempts")
        break
```
